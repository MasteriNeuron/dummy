```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data points for classes A and B
class_A = np.array([(3, 1), (3, 4), (3, 5), (2, 5), (4, 5), (4, 6)])
class_B = np.array([(5, 1), (5, 3), (5, 4), (6, 1), (6, 2), (7, 1), (7, 2)])

# Function to plot data points
def plot_data(class_A, class_B):
    plt.scatter(class_A[:, 0], class_A[:, 1], marker='o', label='Class A')
    plt.scatter(class_B[:, 0], class_B[:, 1], marker='x', label='Class B')

# Define a function to plot the decision boundary
def plot_decision_boundary(w0, w1, w2, title):
    x_range = np.linspace(1, 8, 800)
    y_range = np.linspace(0, 7, 800)
    X, Y = np.meshgrid(x_range, y_range)
    Z = w0 + w1 * X + w2 * Y
    
    plt.contour(X, Y, Z, levels=[0], colors='k', linestyles='dashed')
    plt.title(title)

# Scenario (a): Lasso Regularization on All Parameters
w0_a, w1_a, w2_a = 0.5, -0.5, 0.5
plt.figure(figsize=(12, 4))
plt.subplot(131)
plot_data(class_A, class_B)
plot_decision_boundary(w0_a, w1_a, w2_a, "Lasso on All Parameters")

# Scenario (b): Lasso Regularization on w0 Only
w0_b, w1_b, w2_b = 0, -0.5, 0.5
plt.subplot(132)
plot_data(class_A, class_B)
plot_decision_boundary(w0_b, w1_b, w2_b, "Lasso on w0 Only")

# Scenario (c): Lasso Regularization on w1 Only
w0_c, w1_c, w2_c = 0.5, -0.0000001, 0.5
plt.subplot(133)
plot_data(class_A, class_B)
plot_decision_boundary(w0_c, w1_c, w2_c, "Lasso on w1 Only")

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the data points for classes A and B
class_A = np.array([(3, 1), (3, 4), (3, 5), (2, 5), (4, 5), (4, 6)])
class_B = np.array([(5, 1), (5, 3), (5, 4), (6, 1), (6, 2), (7, 1), (7, 2)])

# Define the regularization parameter (lambda)
lambda_value = 0.1  # You can adjust this value

# Function to calculate sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Function to calculate decision boundary
def decision_boundary(w0, w1, w2, x1, x2):
    z = w0 + w1 * x1 + w2 * x2
    return sigmoid(z)

# Generate data points for the decision boundary
x1 = np.linspace(1, 8, 400)
x2 = np.linspace(1, 7, 350)
X1, X2 = np.meshgrid(x1, x2)
Y = decision_boundary(0, 0, 0, X1, X2)

# Create a 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot data points for class A and class B
ax.scatter(class_A[:, 0], class_A[:, 1], np.zeros_like(class_A[:, 0]), c='b', marker='o', label='Class A')
ax.scatter(class_B[:, 0], class_B[:, 1], np.zeros_like(class_B[:, 0]), c='r', marker='x', label='Class B')

# Plot the decision boundary
ax.plot_surface(X1, X2, Y, cmap='viridis', alpha=0.5, antialiased=True)

# Set labels and legend
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Probability')
ax.set_title('Decision Boundary in 3D')

# Show legend
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define data points for classes A and B
class_A = np.array([(3, 1), (3, 4), (3, 5), (2, 5), (4, 5), (4, 6)])
class_B = np.array([(5, 1), (5, 3), (5, 4), (6, 1), (6, 2), (7, 1), (7, 2)])

# Helper function to plot data points
def plot_data_points(class_A, class_B):
    plt.scatter(class_A[:, 0], class_A[:, 1], marker='o', label='Class A')
    plt.scatter(class_B[:, 0], class_B[:, 1], marker='x', label='Class B')

# Define the decision boundary equations for each case

# (a) Decision boundary for optimal model (Lasso on all parameters)
def decision_boundary_optimal(x):
    w0, w1, w2 = 0.0, 0.0, 0.0  # Replace with actual values
    return (-w0 - w1 * x) / w2

# (b) Decision boundary with Lasso regularization on w0 only
def decision_boundary_regularize_w0(x):
    w0, w1, w2 = 0.0, 0.0, 0.0  # Replace with actual values
    return (-w0 - w1 * x) / w2

# (c) Decision boundary with Lasso regularization on w1 only
def decision_boundary_regularize_w1(x):
    w0, w1, w2 = 0.0, 0.0, 0.0  # Replace with actual values
    return (-w0 - w1 * x) / w2

# Create a range of x values for plotting
x_range = np.linspace(1, 8, 100)

# Create subplots for each case
plt.figure(figsize=(15, 5))

# Plot for (a)
plt.subplot(131)
plot_data_points(class_A, class_B)
plt.plot(x_range, decision_boundary_optimal(x_range), label='Decision Boundary')
plt.title('Decision Boundary (Optimal)')
plt.legend()

# Plot for (b)
plt.subplot(132)
plot_data_points(class_A, class_B)
plt.plot(x_range, decision_boundary_regularize_w0(x_range), label='Decision Boundary')
plt.title('Decision Boundary (Regularize w0)')
plt.legend()

# Plot for (c)
plt.subplot(133)
plot_data_points(class_A, class_B)
plt.plot(x_range, decision_boundary_regularize_w1(x_range), label='Decision Boundary')
plt.title('Decision Boundary (Regularize w1)')
plt.legend()

plt.tight_layout()
plt.show()
```
